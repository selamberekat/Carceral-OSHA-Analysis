##### iii. Principal Component Analysis (PCA):

```python
scaler = StandardScaler() # Standardize the data (important for PCA)
X = scaler.fit_transform(n2)
pca = PCA() # Fit PCA
pca.fit(X)
```

```python
print(pca.explained_variance_ratio_) # Explained Variance Ratio
```

```python
print(pca.components_) # Principal Components
```

```python
# Plot scree plot
plt.figure(figsize=(10, 6))
plt.bar(range(1, len(pca.explained_variance_ratio_) + 1), pca.explained_variance_ratio_, align='center', alpha=0.8)
plt.xticks(range(1, len(pca.explained_variance_ratio_) + 1), ['PC{}'.format(i) for i in range(1, len(pca.explained_variance_ratio_) + 1)])
plt.xlabel('Principal Components')
plt.ylabel('Explained Variance Ratio')
plt.title('Scree Plot')
plt.grid(True)
plt.show()
```

```python
# Determine variable names associated with each principal component
component_names = ['PC{}'.format(i+1) for i in range(len(pca.explained_variance_ratio_))]
variable_names = n2.columns

# Print the association of variables with principal components (sorted in ascending order)
for component, component_name in zip(pca.components_, component_names):
    sorted_indices = np.abs(component).argsort()[::-1]
    sorted_variables = variable_names[sorted_indices]
    sorted_coefficients = component[sorted_indices]
    print(f"{component_name}:")
    
    for variable, coefficient in zip(sorted_variables, sorted_coefficients):
        print(f"{variable}: {coefficient:.4f}")
    print()
```

```python
def calculate_zeros_and_merge(df=new_df):
    total_establishments = df['owner_typef.1'].value_counts().reset_index()
    total_establishments.columns = ['owner_typef.1', 'Total Establishments']

    zeros_owner1 = df.groupby('owner_typef.1')[['nr_violations.1', 'ipavg.1', 'cpavg.1']].apply(lambda x: (x == 0).sum()).reset_index()
    zeros_owner1 = zeros_owner1.merge(total_establishments, on='owner_typef.1', how='left')
    
    union_counts = df.groupby('owner_typef.1')['union_majority.1'].agg(['sum', 'count']).reset_index()
    union_counts.columns = ['owner_typef.1', 'Union Present', 'Total']
    union_counts['Union Absent'] = union_counts['Total'] - union_counts['Union Present']
    
    final_df = zeros_owner1.merge(union_counts[['owner_typef.1', 'Union Present', 'Union Absent']], 
                                  on='owner_typef.1', 
                                  how='left')
    
    # Calculate average reduction
    avg_ip = df.groupby('owner_typef.1')['ipavg.1'].mean().reset_index()
    avg_cp = df.groupby('owner_typef.1')['cpavg.1'].mean().reset_index()
    
    final_df = final_df.merge(avg_ip, on='owner_typef.1', how='left').rename(columns={'ipavg.1': 'Avg Initial Viol Pay'})
    final_df = final_df.merge(avg_cp, on='owner_typef.1', how='left').rename(columns={'cpavg.1': 'Avg Current Viol Pay'})
    
    # Calculate total reduction of violation pay per facility
    total_initial_payment = df.groupby('owner_typef.1')['ipavg.1'].sum()
    total_current_payment = df.groupby('owner_typef.1')['cpavg.1'].sum()
    reduction = (total_initial_payment - total_current_payment).reset_index()
    reduction.columns = ['owner_typef.1', 'Total Reduction of Viol Pay']
    
    final_df = final_df.merge(reduction, on='owner_typef.1', how='left')
    
    # Calculate total violations
    total_violations = df.groupby('owner_typef.1')['nr_violations.1'].sum().reset_index()
    total_violations.columns = ['owner_typef.1', 'Total Violations']
    
    final_df = final_df.merge(total_violations, on='owner_typef.1', how='left')
    
    # Calculate average reduction per violation
    final_df['Average Reduction'] = final_df['Total Reduction of Viol Pay'] / final_df['Total Violations']
    
    column_names = {'owner_typef.1': 'Owner', 
                    'nr_violations.1': 'Nr Violations', 
                    'Total Establishments': 'Total Establishments', 
                    'Union Present': 'Union Present', 
                    'Union Absent': 'Union Absent', 
                    'ipavg.1': 'Count of Initial Viol Pay', 
                    'cpavg.1': 'Count of Current Viol Pay', 
                    'Avg Initial Viol Pay': 'Avg Initial Viol Pay', 
                    'Avg Current Viol Pay': 'Avg Current Viol Pay', 
                    'Total Reduction of Viol Pay': 'Total Reduction',
                    'Total Violations': 'Total Violations',
                    'Average Reduction': 'Average Reduction'}
    return final_df.rename(columns=column_names)
```

```python
# Calculate overall totals for original DataFrame
total_initial_payment = new_df['ipavg.1'].sum()
total_current_payment = new_df['cpavg.1'].sum()
total_difference = total_initial_payment - total_current_payment
average_reduction_per_violation = total_difference / new_df['nr_violations.1'].sum()
print("Original DataFrame (Without Dropping Zero Violations):")
print("Total Initial Pay:", total_initial_payment)
print("Total Current Pay:", total_current_payment)
print("Total Reduction in Violation Penalty: USD", round(total_difference, 2))
print("Average Reduction per Violation: USD", round(average_reduction_per_violation, 2))
print(" ")
zeros_owner_original = calculate_zeros_and_merge(new_df)
renamed_columns = {
    'Owner': 'Facility Owner',
    'Nr Violations': 'Number of Violations',
    'Total Establishments': 'Total Facilities',
    'Union Present': 'Unionized Facilities',
    'Union Absent': 'Non-unionized Facilities',
    'Count of Initial Viol Pay': 'Initial Violations Pay Count',
    'Count of Current Viol Pay': 'Current Violations Pay Count',
    'Avg Initial Viol Pay': 'Average Initial Violation Payment',
    'Avg Current Viol Pay': 'Average Current Violation Payment',
    'Total Reduction': 'Total Violation Payment Reduction',
    'Total Violations': 'Total Number of Violations',
    'Average Reduction': 'Average Payment Reduction per Violation'
}
zeros_owner_original_renamed = zeros_owner_original.rename(columns=renamed_columns)
print(zeros_owner_original_renamed)
print(" ")
```
